```python
########importing libraries
import os
#regrex
import re
#lib
import glob
from datetime import datetime , timedelta, date
import pandas as pd
#importing database access
import warnings
warnings.simplefilter(action='ignore', category=FutureWarning)
#define root here
rootdir = r'C:\Users\ajay.dubey\Documents\Task\Task4_sunshine\Sunsure'
#get a date to use in glob
end_date = date.today() 
start_date = end_date - timedelta(days=10)
#Getting list of dates
def daterange(start_date, end_date):
    for n in range(int((end_date - start_date).days)+1):
        yield start_date + timedelta(n)    
#req. data 
def data_extract(filePath,fileDate):
    df=pd.read_csv(filePath, encoding = 'unicode_escape')
    print(df.columns)
    df=df[['Datetime','GHI Irradiance (w/m?)','Ambient Temperature (?C)','Module Temperature  (?C)','Active Power (kW)']]
    df['Datetime'] = pd.to_datetime(df['Datetime'].astype(str)) - pd.DateOffset(minutes=2)
    df['Datetime'] = df['Datetime'].dt.strftime('%Y-%m-%d %H:%M:%S')
    # for testing : 'ss{0}.json'.format(fileDate),
    return df.to_json(orient='records')
#for generating the json file/output
for currDate in daterange(start_date, end_date):
    cdfname = currDate.strftime('%Y%m%d')    
    listCurrDayFiles = glob.glob(rootdir+('/*{}.csv*'.format(cdfname)))
    #print(listCurrDayFiles)
    for filePath in listCurrDayFiles:
        fileName = os.path.basename(filePath)
        dirname, fileDate = fileName.split(' ')
        print(data_extract(filePath,fileDate))
```
